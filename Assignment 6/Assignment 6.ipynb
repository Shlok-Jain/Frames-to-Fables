Another important point is that, some models will be even larger that the computational resources provided by Kaggle or Google colab, so in that case you will have to load the model in quantized format, which will reduce the size of the model and make it easier to run on limited resources.

Quantization: It is a technique used to reduce the size of a model by converting its weights from floating-point precision to lower precision, for eg. if originally the weights are in 32-bit floating point format, then you may convert them to 16 or 8 or even 4-bit format. This reduces the memory footprint of the model and makes it easier to run on limited resources. However this may also reduce the performance of the model, so you should try to find a balance between the size and performance of the model.

For loading a model in quantized format, there is a llibrary called `bitsandbytes`, for more details on how to use it, you can refer to online resources or ChatGPT, etc.

````python
# Code for Story Generation
# Feel free to use Kaggle or Colab for actually running the code, but paste it here after you are done.
from huggingface_hub import login
login()

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig, StoppingCriteriaList
import os

def generate_story(model_id, captions, model_name_for_file="model", quantized=True, max_new_tokens=1000):
    """
    Generate a story from captions using a given model.

    Args:
        model_id (str): Hugging Face model ID
        captions (list): List of caption strings
        model_name_for_file (str): Short name for saving file
        quantized (bool): Whether to load model in 8-bit using bitsandbytes
        max_new_tokens (int): Maximum tokens to generate

    Returns:
        str: The generated story text
    """

    print(f"\nğŸ”„ Loading model: {model_id}...\n")


    joined_captions = ' '.join(captions)
    prompt = f"""
You are a creative writer. Write a long, detailed story based on the following image captions:
{joined_captions}
Include characters, emotions, setting, and an ending. Make it at least 300 words long.
"""


    tokenizer = AutoTokenizer.from_pretrained(model_id)


    if quantized:
        bnb_config = BitsAndBytesConfig(
            load_in_8bit=True,
            llm_int8_enable_fp32_cpu_offload=True
        )
        model = AutoModelForCausalLM.from_pretrained(
            model_id,
            device_map="auto",
            quantization_config=bnb_config
        )
    else:
        model = AutoModelForCausalLM.from_pretrained(
            model_id,
            device_map="auto"
        )


    generator = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer
    )

    story_output = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        temperature=0.9,
        top_p=0.95,
        do_sample=True,
        repetition_penalty=1.1,
        stopping_criteria=StoppingCriteriaList(),
        eos_token_id=tokenizer.eos_token_id
    )[0]['generated_text']


    os.makedirs("outputs", exist_ok=True)
    file_path = f"outputs/{model_name_for_file}_story.txt"
    with open(file_path, "w") as f:
        f.write(story_output)


    print("\nâœ… === Generated Story === âœ…\n")
    print(story_output)
    print(f"\nğŸ’¾ Saved to: {file_path}")

    return story_output


captions = [
    "A boy stands in a field with a kite.",
    "The kite flies high in the sky.",
    "Dark clouds begin to gather.",
    "Rain starts pouring, and the boy runs home."
]



# Generate story

# model 1
# Model name: Falcon-7B-Instruct
# Link:https://huggingface.co/tiiuae/falcon-7b-instruct
# Code:
generate_story(
    model_id="tiiuae/falcon-7b-instruct",
    captions=captions,
    model_name_for_file="falcon7b",
    quantized=True,
    max_new_tokens=1000
)

# model 2
# Model name: TinyLlama
# Link:https://huggingface.co/TinyLlama/TinyLlama-1.1B-Chat-v1.0
# Code:
generate_story(
    model_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    captions=captions,
    model_name_for_file="tinyllama",
    quantized=True,
    max_new_tokens=1000
)

# model 3
# Model name: OpenChat 3.5
# Link:https://huggingface.co/openchat/openchat-3.5-1210
# Code:
generate_story(
    model_id="openchat/openchat-3.5-1210",
    captions=captions,
    model_name_for_file="openchat",
    quantized=True,
    max_new_tokens=1000
)

# model 4
# Model name: Mistral-7B
# Link:https://huggingface.co/mistralai/Mistral-7B-Instruct-v0.1
# Code:
generate_story(
    model_id="mistralai/Mistral-7B-Instruct-v0.1",
    captions=captions,
    model_name_for_file="mistral7b",
    quantized=True,
    max_new_tokens=1000
)

````

````python
from huggingface_hub import login
login()

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig, StoppingCriteriaList
import os

def generate_story(model_id, captions, model_name_for_file="model", quantized=True, max_new_tokens=1000):
    """
    Generate a story from captions using a given model.

    Args:
        model_id (str): Hugging Face model ID
        captions (list): List of caption strings
        model_name_for_file (str): Short name for saving file
        quantized (bool): Whether to load model in 8-bit using bitsandbytes
        max_new_tokens (int): Maximum tokens to generate

    Returns:
        str: The generated story text
    """

    print(f"\nğŸ”„ Loading model: {model_id}...\n")

    joined_captions = ' '.join(captions)
    prompt = f"""
You are a creative writer. Write a long, detailed story based on the following image captions:
{joined_captions}
Include characters, emotions, setting, and an ending. Make it at least 300 words long.
"""

    tokenizer = AutoTokenizer.from_pretrained(model_id)

    if quantized:
        bnb_config = BitsAndBytesConfig(
            load_in_8bit=True,
            llm_int8_enable_fp32_cpu_offload=True
        )
        model = AutoModelForCausalLM.from_pretrained(
            model_id,
            device_map="auto",
            quantization_config=bnb_config
        )
    else:
        model = AutoModelForCausalLM.from_pretrained(
            model_id,
            device_map="auto"
        )

    generator = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer
    )

    story_output = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        temperature=0.9,
        top_p=0.95,
        do_sample=True,
        repetition_penalty=1.1,
        stopping_criteria=StoppingCriteriaList(),
        eos_token_id=tokenizer.eos_token_id
    )[0]['generated_text']

    os.makedirs("outputs", exist_ok=True)
    file_path = f"outputs/{model_name_for_file}_story.txt"
    with open(file_path, "w") as f:
        f.write(story_output)

    print("\nâœ… === Generated Story === âœ…\n")
    print(story_output)
    print(f"\nğŸ’¾ Saved to: {file_path}")

    return story_output

# Caption describing the image
captions = [
    "A cheerful boy runs through a golden field holding a colorful kite that dances in the clear blue sky.",
    "Suddenly the weather changes and dark clouds gather.",
    "it starts raining heavily.",
    "the boy starts running."
]

# === MODEL 1: GPT-2 ===
generate_story(
    model_id="gpt2",
    captions=captions,
    model_name_for_file="gpt2_kite_story",
    quantized=False,          # GPT-2 is small; quantization unnecessary
    max_new_tokens=1000
)

# === MODEL 2: GPT-Neo 1.3B ===
generate_story(
    model_id="EleutherAI/gpt-neo-1.3B",
    captions=captions,
    model_name_for_file="gptneo1.3b_kite_story",
    quantized=True,           # Enable quantization to save memory
    max_new_tokens=1000
)

# === MODEL 3: GPT-J 6B ===
generate_story(
    model_id="EleutherAI/gpt-j-6B",
    captions=captions,
    model_name_for_file="gptj6b_kite_story",
    quantized=True,           # GPT-J is big, quantization recommended
    max_new_tokens=1000
)
````
